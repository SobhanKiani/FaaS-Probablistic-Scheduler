```python
import numpy as np
import matplotlib.pyplot as plt 
import matplotlib.patches as mpatches
import pandas as pd 

plt.style.use('bmh')
bar_chart_width = 0.2
```

```python
columns = ['Time', 'Memory', 'Cs Time', 'Cs Count']

cs = pd.read_csv('./cold-start.txt', header=None)
probable = pd.read_csv('./mlp.txt', header=None)
optimal = pd.read_csv('./warming-all.txt', header=None)
optimal_parameter = pd.read_csv('./optimal.txt', header=None)

cs.columns = columns
probable.columns = columns
optimal.columns = columns
optimal_parameter.columns = columns
```

```python
cs.shape, probable.shape, optimal.shape, optimal_parameter.shape
```

```python
cs_time_mean = cs['Time'].mean()
cs_mem_mean = cs['Memory'].mean()
cs_mean_cs_time = cs['Cs Time'].mean()
cs_mean_cs_count = cs['Cs Count'].mean()

probable_time_mean = probable['Time'].mean()
probable_mem_mean = probable['Memory'].mean()
probable_mean_cs_time = probable['Cs Time'].mean()
probable_mean_cs_count = probable['Cs Count'].mean()

optimal_time_mean = optimal['Time'].mean()
optimal_mem_mean = optimal['Memory'].mean()
optimal_mean_cs_time = optimal['Cs Time'].mean()
optimal_mean_cs_count = optimal['Cs Count'].mean()

optimal_parameter_time_mean = optimal_parameter['Time'].mean()
optimal_parameter_mem_mean = optimal_parameter['Memory'].mean()
optimal_parameter_mean_cs_time = optimal_parameter['Cs Time'].mean()
optimal_parameter_mean_cs_count = optimal_parameter['Cs Count'].mean()

cs_time_std = cs['Time'].std()
cs_mem_std = cs['Memory'].std()
cs_std_cs_time = cs['Cs Time'].std()
cs_std_cs_count = cs['Cs Count'].std()

probable_time_std = probable['Time'].std()
probable_mem_std = probable['Memory'].std()
probable_std_cs_time = probable['Cs Time'].std()
probable_std_cs_count = probable['Cs Count'].std()

optimal_time_std = optimal['Time'].std()
optimal_mem_std = optimal['Memory'].std()
optimal_std_cs_time = optimal['Cs Time'].std()
optimal_std_cs_count = optimal['Cs Count'].std()

optimal_parameter_time_std = optimal_parameter['Time'].std()
optimal_parameter_mem_std = optimal_parameter['Memory'].std()
optimal_parameter_std_cs_time = optimal_parameter['Cs Time'].std()
optimal_parameter_std_cs_count = optimal_parameter['Cs Count'].std()
```

```python
time_mean_data = [cs_time_mean, probable_time_mean,
                  optimal_time_mean, optimal_parameter_time_mean]
time_std_data = [cs_time_std, probable_time_std,
                 optimal_time_std, optimal_parameter_time_std]

mem_mean_data = [cs_mem_mean, probable_mem_mean,
                 optimal_mem_mean, optimal_parameter_mem_mean]
mem_std_data = [cs_mem_std, probable_mem_std,
                optimal_mem_std, optimal_parameter_mem_std]

cs_time_mean_data = [cs_mean_cs_time, probable_mean_cs_time,
                     optimal_mean_cs_time, optimal_parameter_mean_cs_time]
cs_time_std_data = [cs_std_cs_time, probable_std_cs_time,
                     optimal_std_cs_time, optimal_parameter_std_cs_time]

cs_count_mean_data = [cs_mean_cs_count, probable_mean_cs_count,
                     optimal_mean_cs_count, optimal_parameter_mean_cs_count]
cs_count_std_data = [cs_std_cs_count, probable_std_cs_count,
                     optimal_std_cs_count, optimal_parameter_std_cs_count]
```

# Plotting The Bars Charts

```python
colors = ['#32a852', '#a84343', '#5124b3', '#3856ff']
labels = ['No Solutions', 'Probabilistic', 'Optimal', 'Optimal Parameteres']
```

```python
labels = ['No Solutions', 'Probabilistic', 'Optimal', 'Optimal Parameteres']

# colors = plt.cm.viridis(np.linspace(0, 1, len(data)))

plt.bar(labels, time_mean_data, bar_chart_width, color=colors)
plt.title('Mean Total Execution Time of the DAG', fontdict={'family': 'Arial', 'size': 12})
plt.xlabel('Approach', fontdict={'family': 'Arial', 'size': 12})
plt.ylabel('Time (In Seconds)', fontdict={'family': 'Arial', 'size': 12})
plt.xticks(fontname='Arial', fontsize=10, rotation=45)
plt.yticks(fontname='Arial', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```

```python

# colors = plt.cm.viridis(np.linspace(0, 1, len(data)))

plt.bar(labels, mem_mean_data, bar_chart_width, color=colors)
plt.title('Mean Total Memory of the DAG', fontdict={'family': 'Arial', 'size': 12})
plt.xlabel('Approach', fontdict={'family': 'Arial', 'size': 12})
plt.ylabel('Size in MB', fontdict={'family': 'Arial', 'size': 12})
plt.xticks(fontname='Arial', fontsize=10, rotation=45)
plt.yticks(fontname='Arial', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```

```python
plt.bar(labels, cs_time_mean_data, bar_chart_width, color=colors)
plt.title('Mean Total Cold Start Time of the DAG', fontdict={'family': 'Arial', 'size': 12})
plt.xlabel('Approach', fontdict={'family': 'Arial', 'size': 12})
plt.ylabel('Time (In Seconds)', fontdict={'family': 'Arial', 'size': 12})
plt.xticks(fontname='Arial', fontsize=10, rotation=45)
plt.yticks(fontname='Arial', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```

```python
plt.bar(labels, cs_count_mean_data, bar_chart_width, color=colors)
plt.title('Mean Total Cold Start Counts of the DAG',
          fontdict={'family': 'Arial', 'size': 12})
plt.xlabel('Approach', fontdict={'family': 'Arial', 'size': 12})
plt.ylabel('Time (In Seconds)', fontdict={'family': 'Arial', 'size': 12})
plt.xticks(fontname='Arial', fontsize=10, rotation=45)
plt.yticks(fontname='Arial', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend()

plt.show()
```

# Line Graph: Memory And Time Together

```python
# colors = plt.cm.viridis(np.linspace(0, 1, len(data)))

plt.scatter(cs['Memory'], cs['Time'], c=colors[0], )
plt.scatter(probable['Memory'], probable['Time'], c=colors[1], marker='^')
plt.scatter(optimal['Memory'], optimal['Time'], c=colors[2], marker='x')
plt.scatter(optimal_parameter['Memory'], optimal_parameter['Time'], c=colors[3], marker='x')

plt.title('Memory Usage Vs. Ex Time', fontdict={'family': 'Arial', 'size': 12})
plt.xlabel('Memory Usage (In MB)', fontdict={'family': 'Arial', 'size': 12})
plt.ylabel('Time (In Seconds)', fontdict={'family': 'Arial', 'size': 12})
plt.xticks(fontname='Arial', fontsize=10, rotation=45)
plt.yticks(fontname='Arial', fontsize=10)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(['Without Any Solutions', 'Probable', 'Warming All', 'Optimal'])

plt.show()
```

# Plotting Mean And Stds

```python
x = range(len(labels))

fig, ax1 = plt.subplots()
ax1.bar(x, time_mean_data, bar_chart_width, label='Time Mean', color=colors)
ax1.errorbar(x, time_mean_data, yerr=time_std_data, fmt='o', capsize=5)
ax1.errorbar(x, time_mean_data, yerr=time_std_data, capsize=5)

# ax2 = ax1.twinx()
# ax2.bar(x, mem_mean_data, width, label='Memory Mean', alpha=0.5)
# ax2.errorbar(x, mem_mean_data, yerr=mem_std_data, fmt='o', capsize=5)

ax1.set_ylabel('Time')
# ax2.set_ylabel('Memory')

plt.xticks(x, labels)
plt.xlabel('Groups')
plt.title('Execution Time Means with Standard Deviations')

legend_patches = [mpatches.Patch(color=color, label=label) for color, label in zip(colors, labels)]
plt.legend(handles=legend_patches)
plt.tight_layout()
plt.show()
```

```python
x = range(len(labels))

fig, ax1 = plt.subplots()
ax1.bar(x, mem_mean_data, bar_chart_width, label='Time Mean', color=colors)
ax1.errorbar(x, mem_mean_data, yerr=mem_std_data, fmt='o', capsize=5)
ax1.errorbar(x, mem_mean_data, yerr=mem_std_data, capsize=5)

ax1.set_ylabel('Memory')

plt.xticks(x, labels)
plt.xlabel('Groups')
plt.title('Memory Means with Standard Deviations')

legend_patches = [mpatches.Patch(color=color, label=label) for color, label in zip(colors, labels)]
plt.legend(handles=legend_patches)
plt.tight_layout()
plt.show()
```
